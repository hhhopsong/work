```python
import numpy as np
import scipy.signal
import matplotlib.pyplot as plt
# 加载和预处理数据
data = np.load("D:\PyFile\paper1\OLS35.npy")  # 示例数据，可根据实际情况修改
data = np.array(data, dtype=complex)
# 计算功率谱密度（PSD）
frequencies, psd = scipy.signal.periodogram(data)
# 计算平均功率频率
average_power_frequency = np.sum(psd) / len(psd)
print('Average power frequency:', average_power_frequency)
# 绘制功率谱图
plt.plot(1/frequencies, psd)
plt.xlabel('Frequency')
plt.ylabel('Power Spectral Density')
plt.show()
```
